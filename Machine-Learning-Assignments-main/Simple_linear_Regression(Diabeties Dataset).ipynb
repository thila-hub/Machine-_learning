```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn import datasets
from sklearn.linear_model import LinearRegression  # Correctly import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score

# Load the diabetes dataset
diabetes = datasets.load_diabetes()

# Display feature names and the first five rows of data
print("Feature names:", diabetes['feature_names'])
print("First five rows of data:\n", diabetes['data'][:5])

# Define the independent variable (feature) and the dependent variable (target)
X = diabetes.data[:, np.newaxis, 2]  # Use only one feature
y = diabetes.target

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=1/3, random_state=0)

# Create a linear regression model and train it
regressor = LinearRegression()
regressor.fit(X_train, y_train)

# Make predictions using the training and testing sets
y_train_pred = regressor.predict(X_train)
y_test_pred = regressor.predict(X_test)

# Plot training data with the regression line
plt.scatter(X_train, y_train, color="green", label="Training data")
plt.plot(X_train, y_train_pred, color="red", label="Regression line")
plt.title("Linear Regression on Diabetes Dataset (Training Set)")
plt.xlabel("Feature")
plt.ylabel("Target")
plt.legend()
plt.show()

# Plot testing data with the regression line
plt.scatter(X_test, y_test, color="blue", label="Testing data")
plt.plot(X_train, y_train_pred, color="red", label="Regression line (Training)")
plt.title("Linear Regression on Diabetes Dataset (Testing Set)")
plt.xlabel("Feature")
plt.ylabel("Target")
plt.legend()
plt.show()

# Display the coefficients and intercept of the model
print('Coefficients:', regressor.coef_)
print('Intercept:', regressor.intercept_)

# Calculate and display the mean squared error and R^2 score
print('Mean Squared Error:', mean_squared_error(y_test, y_test_pred))
print('Coefficient of Determination (R^2 Score):', r2_score(y_test, y_test_pred))
```
